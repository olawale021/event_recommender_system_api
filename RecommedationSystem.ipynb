```python
import pandas as pd
import numpy as np
from scipy.io import arff
from surprise import Dataset, Reader, SVD, KNNWithMeans, accuracy
from surprise.model_selection import train_test_split
from sklearn.model_selection import train_test_split as sklearn_train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Model
from keras.layers import Input, Embedding, Flatten, Dense, Concatenate, Dropout
from keras.optimizers import Adam
import tensorflow as tf

# Data Loading Function
def load_and_preprocess_data():
   
    user_behavior = pd.DataFrame(arff.loadarff('cleaned_user_behaviour_new.arff')[0])
    # Load additional data from CSV files
    events = pd.read_csv('events.csv')
    users = pd.read_csv('users.csv')
    tickets = pd.read_csv('tickets.csv')
    preferred_categories = pd.read_csv('user_preferred_categories.csv')
    friends = pd.read_csv('friends.csv')

    # Decode byte strings in object columns to regular strings
    for col in user_behavior.select_dtypes([object]):
        user_behavior[col] = user_behavior[col].str.decode('utf-8')

    # Ensure user_id and event_id columns are integers for consistency
    user_behavior['user_id'] = user_behavior['user_id'].astype(int)
    user_behavior['event_id'] = user_behavior['event_id'].astype(int)

    return user_behavior, events, users, tickets, preferred_categories, friends

# Feature Engineering Function
def create_user_features(user_behavior, events, users, tickets, preferred_categories, friends):
    df = user_behavior.copy()
    
    # Assign weights to interaction types
    interaction_weights = {
        'browse': 2.5,
        'view': 3.5,
        'purchase': 5.0,
        'add_to_wishlist': 4.0,
        'remove_from_wishlist': 0,
        'view_friend_profile': 2.5
    }
    df['interaction_value'] = df['interaction_type'].map(interaction_weights)

    # Standardize session lengths
    df['session_score'] = StandardScaler().fit_transform(df[['session_length']])

    # Calculate a weighted final score
    df['final_score'] = (
        df['interaction_value'] * 0.6 +
        df['session_score'] * 0.4
    )

    # Normalize the final_score to range [0, 1]
    df['final_score'] = MinMaxScaler().fit_transform(df[['final_score']])

    return df


# 3. Preprocess IDs for Compatibility
def preprocess_ids(df):
    """Re-index user_id and event_id to start from 0 for embedding compatibility."""
    # Ensure all IDs are consistent and unique
    df = df.copy()  # Avoid modifying the original dataframe

    unique_users = df['user_id'].unique()
    unique_events = df['event_id'].unique()

    user_id_map = {user_id: idx for idx, user_id in enumerate(unique_users)}
    event_id_map = {event_id: idx for idx, event_id in enumerate(unique_events)}

    # Map IDs
    df['user_id'] = df['user_id'].map(user_id_map)
    df['event_id'] = df['event_id'].map(event_id_map)

    # Drop rows with unmapped (NaN) IDs
    df = df.dropna(subset=['user_id', 'event_id'])

    return df, len(unique_users), len(unique_events)


#  Prepare Training Data for Surprise Models
def prepare_surprise_data(df):
    """Prepare data for Surprise recommendation models."""
    reader = Reader(rating_scale=(0, 1))
    data = Dataset.load_from_df(df[['user_id', 'event_id', 'final_score']], reader)
    return train_test_split(data, test_size=0.2, random_state=42)

#  Train NCF Model
def train_ncf(df):
    """Train a Neural Collaborative Filtering (NCF) model."""
    df, n_users, n_events = preprocess_ids(df)

    user_input = Input(shape=(1,), name='user_input')
    event_input = Input(shape=(1,), name='event_input')

    user_embedding = Embedding(input_dim=n_users, output_dim=50)(user_input)
    event_embedding = Embedding(input_dim=n_events, output_dim=50)(event_input)

    user_vector = Flatten()(user_embedding)
    event_vector = Flatten()(event_embedding)

    merged_vector = Concatenate()([user_vector, event_vector])

    dense = Dense(128, activation='relu')(merged_vector)
    dense = Dropout(0.2)(dense)
    dense = Dense(64, activation='relu')(dense)
    dense = Dense(32, activation='relu')(dense)
    output = Dense(1, activation='sigmoid')(dense)

    model = Model(inputs=[user_input, event_input], outputs=output)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

    train_df, val_df = sklearn_train_test_split(df, test_size=0.2, random_state=42)
    X_train = [train_df['user_id'].values, train_df['event_id'].values]
    y_train = train_df['final_score'].values
    X_val = [val_df['user_id'].values, val_df['event_id'].values]
    y_val = val_df['final_score'].values

    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=64, verbose=1)
    return model, n_users, n_events


#  Train MLP Model
def train_mlp(df):
    """Train a Multi-Layer Perceptron (MLP) model."""
    # Re-index IDs and determine the total number of users and events
    df, n_users, n_events = preprocess_ids(df)

    # Define the inputs for users and events
    user_input = Input(shape=(1,), name='user_input')
    event_input = Input(shape=(1,), name='event_input')

    # Define the embedding layers
    user_embedding = Embedding(input_dim=n_users, output_dim=50, name='user_embedding')(user_input)
    event_embedding = Embedding(input_dim=n_events, output_dim=50, name='event_embedding')(event_input)

    # Flatten the embeddings
    user_vector = Flatten(name='flatten_user_embedding')(user_embedding)
    event_vector = Flatten(name='flatten_event_embedding')(event_embedding)

    # Concatenate the user and event embeddings
    merged_vector = Concatenate(name='concatenate_user_event')([user_vector, event_vector])

    # Add dense layers
    dense = Dense(128, activation='relu', name='dense_128')(merged_vector)
    dense = Dropout(0.2, name='dropout_128')(dense)
    dense = Dense(64, activation='relu', name='dense_64')(dense)
    dense = Dense(32, activation='relu', name='dense_32')(dense)
    output = Dense(1, activation='sigmoid', name='output_layer')(dense)

    # Compile the model
    model = Model(inputs=[user_input, event_input], outputs=output)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

    # Split the data into training and validation sets
    train_df, val_df = sklearn_train_test_split(df, test_size=0.2, random_state=42)
    X_train = [train_df['user_id'].values, train_df['event_id'].values]
    y_train = train_df['final_score'].values
    X_val = [val_df['user_id'].values, val_df['event_id'].values]
    y_val = val_df['final_score'].values

    # Train the model
    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=64, verbose=1)

    # Return the trained model and metadata
    return model, n_users, n_events


from sklearn.metrics import precision_score, recall_score

# 7. Evaluate Model Precision, Recall, and RMSE
def evaluate_model(predictions, threshold=0.5):
    """Evaluate precision, recall, and RMSE for a model's predictions."""
    y_true = np.array([pred.r_ui for pred in predictions])
    y_pred = np.array([pred.est for pred in predictions])

    # RMSE
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

    # Classification threshold for precision and recall
    y_true_binary = (y_true >= threshold).astype(int)
    y_pred_binary = (y_pred >= threshold).astype(int)

    # Precision and Recall
    precision = precision_score(y_true_binary, y_pred_binary, zero_division=1)
    recall = recall_score(y_true_binary, y_pred_binary, zero_division=1)

    return precision, recall, rmse

# 8. Evaluate Neural Models
def evaluate_neural_model(model, df, threshold=0.5, n_users=None, n_events=None):
    """Evaluate a neural model's precision, recall, and RMSE."""
    # Ensure IDs are within range
    if n_users is None or n_events is None:
        raise ValueError("n_users and n_events must be provided to validate IDs.")

    df = df[(df['user_id'] < n_users) & (df['event_id'] < n_events)]  # Filter valid IDs

    X = [df['user_id'].values, df['event_id'].values]
    y_true = df['final_score'].values
    y_pred = model.predict(X).flatten()

    # RMSE
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

    # Classification threshold for precision and recall
    y_true_binary = (y_true >= threshold).astype(int)
    y_pred_binary = (y_pred >= threshold).astype(int)

    # Precision and Recall
    precision = precision_score(y_true_binary, y_pred_binary, zero_division=1)
    recall = recall_score(y_true_binary, y_pred_binary, zero_division=1)

    return precision, recall, rmse


def calculate_weights(rmse_values):
    """Calculate dynamic weights for models based on their RMSE."""
    # Use the inverse of RMSE to calculate weights
    inv_rmse = 1 / np.array(rmse_values)
    weights = inv_rmse / np.sum(inv_rmse)
    return weights


def hybrid_recommendations(user_id, df, svd_model, knn_model, ncf_model, mlp_model, events, preferred_categories_df, weights, category_boost=0.1):

    # Get embedding dimensions from model layers
    user_embedding_layer = [layer for layer in ncf_model.layers if isinstance(layer, tf.keras.layers.Embedding)][0]
    event_embedding_layer = [layer for layer in ncf_model.layers if isinstance(layer, tf.keras.layers.Embedding)][1]
    
    # Ensure user_id is within bounds
    if user_id >= user_embedding_layer.input_dim:
        return pd.DataFrame(columns=['event_id', 'hybrid_score'])
    
    # Get purchased events
    purchased_events = df[(df['user_id'] == user_id) & 
                         (df['interaction_type'] == 'purchase')]['event_id'].unique()
    
    # Get user preferences
    user_prefs = preferred_categories_df[preferred_categories_df['user_id'] == user_id]
    user_top_categories = set(user_prefs['category_id'].unique())
    
    predictions = []
    valid_events = events[events['event_id'] < event_embedding_layer.input_dim]['event_id'].unique()
    
    for event_id in valid_events:
        if event_id in purchased_events:
            continue
            
        try:
            # Get predictions from each model
            svd_pred = svd_model.predict(user_id, event_id).est
            knn_pred = knn_model.predict(user_id, event_id).est
            
            # Neural model predictions with proper ID handling
            if user_id < user_embedding_layer.input_dim and event_id < event_embedding_layer.input_dim:
                ncf_pred = ncf_model.predict([np.array([user_id]), np.array([event_id])])[0][0]
                mlp_pred = mlp_model.predict([np.array([user_id]), np.array([event_id])])[0][0]
            else:
                ncf_pred = 0
                mlp_pred = 0
            
            # Combine scores with proper weighting
            hybrid_score = (weights[0] * svd_pred + 
                          weights[1] * knn_pred +
                          weights[2] * ncf_pred + 
                          weights[3] * mlp_pred)
            
            # Apply category boost during initial scoring
            event_category = events[events['event_id'] == event_id]['category_id'].iloc[0]
            if event_category in user_top_categories:
                hybrid_score += category_boost
                
            predictions.append({
                'event_id': event_id,
                'hybrid_score': hybrid_score,
                'true_score': df[(df['user_id'] == user_id) & 
                                (df['event_id'] == event_id)]['final_score'].mean()
            })
            
        except Exception as e:
            continue
            
    return pd.DataFrame(predictions)

def evaluate_hybrid_model(hybrid_recommendations, df, threshold=0.5):
    """Evaluate hybrid model with aligned predictions."""
    if hybrid_recommendations.empty:
        return 0, 0, float('inf')
        
    # Get true scores for recommended events
    merged_df = hybrid_recommendations.merge(
        df[['event_id', 'final_score']].groupby('event_id')['final_score'].mean().reset_index(),
        on='event_id',
        how='inner'
    )
    
    if merged_df.empty:
        return 0, 0, float('inf')
        
    y_true = merged_df['final_score'].values
    y_pred = merged_df['hybrid_score'].values
    
    # Calculate metrics
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
    y_true_binary = (y_true >= threshold).astype(int)
    y_pred_binary = (y_pred >= threshold).astype(int)
    
    precision = precision_score(y_true_binary, y_pred_binary, zero_division=1)
    recall = recall_score(y_true_binary, y_pred_binary, zero_division=1)
    
    return precision, recall, rmse

def get_neural_predictions(model, user_id, events, user_top_categories, category_boost, n_users, n_events, n_recommendations):
    predictions = []
    for event_id in events['event_id'].unique():
        if user_id < n_users and event_id < n_events:
            try:
                # Get the base score for the event
                base_score = model.predict([np.array([user_id]), np.array([event_id])])[0][0]
                
                # Get the event category
                event_category = events[events['event_id'] == event_id]['category_id'].iloc[0]
                
                # Apply category boost if the event's category is in the user's top categories
                if event_category in user_top_categories:
                    base_score += category_boost
                
                # Store the event ID and boosted score
                predictions.append((event_id, base_score))
            except:
                continue
                
    # Sort the predictions based on the score in descending order and take the top N
    return pd.DataFrame(sorted(predictions, key=lambda x: x[1], reverse=True)[:n_recommendations],
                       columns=['event_id', 'score'])


def print_model_recommendations(user_id, df, svd_model, knn_model, ncf_model, mlp_model, events, hybrid_recs, preferred_categories, n_users=None, n_events=None, n_recommendations=10, category_boost=0.1):
    """Print recommendations from each model for a given user with category information"""

    # Get user preferences and top categories
    user_prefs = preferred_categories[preferred_categories['user_id'] == user_id]
    user_top_categories = set(user_prefs['category_id'].unique())
    
    # Helper function to get and format recommendations
    def format_recommendations(recs):
        return recs.merge(events[['event_id', 'event_name', 'category_id']], on='event_id')[['event_name', 'category_id', 'score']]
    
    # SVD Recommendations
    svd_predictions = [(event_id, svd_model.predict(user_id, event_id).est) 
                      for event_id in events['event_id'].unique() if not pd.isna(event_id)]
    
    # Apply category boost before sorting SVD recommendations
    svd_recs = pd.DataFrame(svd_predictions, columns=['event_id', 'score'])
    for i, row in svd_recs.iterrows():
        event_id = row['event_id']
        event_category = events[events['event_id'] == event_id]['category_id'].iloc[0]
        if event_category in user_top_categories:
            svd_recs.at[i, 'score'] += category_boost

    # Sort by score and take top recommendations
    svd_recs = svd_recs.sort_values(by='score', ascending=False).head(n_recommendations)

    # KNN Recommendations
    knn_predictions = [(event_id, knn_model.predict(user_id, event_id).est) 
                      for event_id in events['event_id'].unique() if not pd.isna(event_id)]
    
    # Apply category boost before sorting KNN recommendations
    knn_recs = pd.DataFrame(knn_predictions, columns=['event_id', 'score'])
    for i, row in knn_recs.iterrows():
        event_id = row['event_id']
        event_category = events[events['event_id'] == event_id]['category_id'].iloc[0]
        if event_category in user_top_categories:
            knn_recs.at[i, 'score'] += category_boost

    # Sort by score and take top recommendations
    knn_recs = knn_recs.sort_values(by='score', ascending=False).head(n_recommendations)

    # Neural Model Recommendations (NCF and MLP) - Use the updated function
    ncf_recs = get_neural_predictions(ncf_model, user_id, events, user_top_categories, category_boost, n_users, n_events, n_recommendations)
    mlp_recs = get_neural_predictions(mlp_model, user_id, events, user_top_categories, category_boost, n_users, n_events, n_recommendations)

    # Hybrid Recommendations (already calculated)
    hybrid_top = hybrid_recs.nlargest(n_recommendations, 'hybrid_score')[['event_id', 'hybrid_score']].rename(columns={'hybrid_score': 'score'})
    # Apply category boost before sorting Hybrid recommendations
    for i, row in hybrid_top.iterrows():
        event_id = row['event_id']
        event_category = events[events['event_id'] == event_id]['category_id'].iloc[0]
        if event_category in user_top_categories:
            hybrid_top.at[i, 'score'] += category_boost
    
    # Format recommendations and print
    print(f"\nTop {n_recommendations} Recommendations for User {user_id}:")
    
    print("\nSVD Recommendations:")
    print(format_recommendations(svd_recs))
    
    print("\nKNN Recommendations:")
    print(format_recommendations(knn_recs))
    
    print("\nNCF Recommendations:")
    print(format_recommendations(ncf_recs))
    
    print("\nMLP Recommendations:")
    print(format_recommendations(mlp_recs))
    
    print("\nHybrid Recommendations:")
    print(format_recommendations(hybrid_top))
    
    return svd_recs, knn_recs, ncf_recs, mlp_recs, hybrid_top

    
def main():
    # Load and preprocess data
    print("Loading data...")
    user_behavior, events, users, tickets, preferred_categories, friends = load_and_preprocess_data()
    
    print("Engineering features...")
    df_engineered = create_user_features(user_behavior, events, users, tickets, preferred_categories, friends)

    # Prepare Surprise data
    print("Preparing training data for Surprise models...")
    trainset, testset = prepare_surprise_data(df_engineered)

    # Train Surprise models
    print("Training SVD model...")
    svd_model = SVD(n_factors=50, n_epochs=20, lr_all=0.005, reg_all=0.02, random_state=42)
    svd_model.fit(trainset)
    svd_predictions = svd_model.test(testset)

    print("Training KNN model...")
    knn_model = KNNWithMeans(k=20, sim_options={'name': 'pearson_baseline', 'user_based': True})
    knn_model.fit(trainset)
    knn_predictions = knn_model.test(testset)

    # Evaluate Surprise models
    print("Evaluating Surprise models...")
    svd_precision, svd_recall, svd_rmse = evaluate_model(svd_predictions)
    knn_precision, knn_recall, knn_rmse = evaluate_model(knn_predictions)

   # Train Neural Collaborative Filtering (NCF) model
    print("Training NCF model...")
    ncf_model, ncf_n_users, ncf_n_events = train_ncf(df_engineered)
    
    # Evaluate NCF model
    print("Evaluating NCF model...")
    ncf_precision, ncf_recall, ncf_rmse = evaluate_neural_model(ncf_model, df_engineered, n_users=ncf_n_users, n_events=ncf_n_events)
    
    # Train Multi-Layer Perceptron (MLP) model
    print("Training MLP model...")
    mlp_model, mlp_n_users, mlp_n_events = train_mlp(df_engineered)
    
    # Evaluate MLP model
    print("Evaluating MLP model...")
    mlp_precision, mlp_recall, mlp_rmse = evaluate_neural_model(mlp_model, df_engineered, n_users=mlp_n_users, n_events=mlp_n_events)
    

    
 # Calculate dynamic weights
    print("Calculating dynamic weights...")
    rmse_values = [svd_rmse, knn_rmse, ncf_rmse, mlp_rmse]
    weights = calculate_weights(rmse_values)

    # Generate hybrid recommendations
    print("Generating hybrid recommendations...")
    hybrid_recs = hybrid_recommendations(
        user_id=3,  # Example user_id
        df=df_engineered,
        svd_model=svd_model,
        knn_model=knn_model,
        ncf_model=ncf_model,
        mlp_model=mlp_model,
        events=events,
        preferred_categories_df=preferred_categories,
        weights=weights,
        category_boost=0.1
    )

    # Evaluate hybrid model
    print("Evaluating hybrid model...")
    hybrid_precision, hybrid_recall, hybrid_rmse = evaluate_hybrid_model(hybrid_recs, df_engineered)

    # Print evaluation results
    print("\nModel Evaluation Results:")
    print(f"SVD - Precision: {svd_precision:.4f}, Recall: {svd_recall:.4f}, RMSE: {svd_rmse:.4f}")
    print(f"KNN - Precision: {knn_precision:.4f}, Recall: {knn_recall:.4f}, RMSE: {knn_rmse:.4f}")
    print(f"NCF - Precision: {ncf_precision:.4f}, Recall: {ncf_recall:.4f}, RMSE: {ncf_rmse:.4f}")
    print(f"MLP - Precision: {mlp_precision:.4f}, Recall: {mlp_recall:.4f}, RMSE: {mlp_rmse:.4f}")
    print(f"Hybrid - Precision: {hybrid_precision:.4f}, Recall: {hybrid_recall:.4f}, RMSE: {hybrid_rmse:.4f}")

    # Visualize RMSE comparison
    models = ['SVD', 'KNN', 'NCF', 'MLP', 'Hybrid']
    rmse_values.append(hybrid_rmse)

    plt.figure(figsize=(10, 5))
    sns.barplot(x=models, y=rmse_values, palette='viridis')
    plt.title("RMSE Comparison for Recommendation Models")
    plt.xlabel("Model")
    plt.ylabel("RMSE")
    plt.show()

     # Visualize Precision and Recall
    precision_values = [svd_precision, knn_precision, ncf_precision, mlp_precision, hybrid_precision]
    recall_values = [svd_recall, knn_recall, ncf_recall, mlp_recall, hybrid_recall]

    # Precision Plot
    plt.figure(figsize=(10, 5))
    sns.barplot(x=models, y=precision_values, palette='coolwarm')
    plt.title("Precision Comparison for Recommendation Models")
    plt.xlabel("Model")
    plt.ylabel("Precision")
    plt.show()

    # Recall Plot
    plt.figure(figsize=(10, 5))
    sns.barplot(x=models, y=recall_values, palette='coolwarm')
    plt.title("Recall Comparison for Recommendation Models")
    plt.xlabel("Model")
    plt.ylabel("Recall")
    plt.show()


    print("\nGenerating individual model recommendations...")
    svd_recs, knn_recs, ncf_recs, mlp_recs, hybrid_recs = print_model_recommendations(
    user_id=4,
    df=df_engineered,
    svd_model=svd_model,
    knn_model=knn_model, 
    ncf_model=ncf_model,
    mlp_model=mlp_model,
    events=events,
    preferred_categories=preferred_categories,
    hybrid_recs=hybrid_recs,
    n_users=ncf_n_users,
    n_events=ncf_n_events,
    category_boost=0.1
)

# Saving models and hybrid weights
    import os
    import pickle

    # Create a directory to store the models if it doesn't exist
    model_dir = 'saved_models'
    if not os.path.exists(model_dir):
        os.makedirs(model_dir)

    # Save Surprise models (SVD and KNN)
    with open(os.path.join(model_dir, 'svd_model.pkl'), 'wb') as f:
        pickle.dump(svd_model, f)

    with open(os.path.join(model_dir, 'knn_model.pkl'), 'wb') as f:
        pickle.dump(knn_model, f)

    # Save Keras models (NCF and MLP)
    ncf_model.save(os.path.join(model_dir, 'ncf_model.h5'))
    mlp_model.save(os.path.join(model_dir, 'mlp_model.h5'))

    # Save hybrid model's dynamic weights
    weights_file = os.path.join(model_dir, 'hybrid_weights.pkl')
    with open(weights_file, 'wb') as f:
        pickle.dump(weights, f)

    print("Models and hybrid weights have been saved successfully.")


if __name__ == "__main__":
    main()
```
